```python
import os
import xlrd 
import csv

eq_header = ["Site Number",
                "Tower Asset Number",
                "Project Number",
                "Project Name",
                "Customer Equipment Number",
                "Customer Equipment Name",
                "Portfolio Country",
                "Status",
               
              
                "Customer Name",
             
                "Customer Equipment Type",
             
                "antenna instalation status",
                "antenna equipment type",
                "antenna mount type",  
                "antenna make",  
                "antenna model",  
                "antenna height",             
                "antenna width",  
                "antenna depth",
                "azimuth",
                "feeder/Line qty",
                "feeder/Line type"]


ground_data = ''
tower_data = ''

ground = []
antenna = []
customer = []

site_number = ''
customer_name = ''

#Root directory
rootdir = 'C:/OTM'


# run the vbs scipt to encode the excel files
os.system("MyScript.vbs 1")


# Reading form def source and writing in customer_eg file
with open('customer_eq.csv',  'w', newline='') as csvfile:
    cust_eq = csv.writer(csvfile, delimiter=',',escapechar = ' ', quoting=csv.QUOTE_NONE) 
    cust_eq.writerow(eq_header)
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            filepath = subdir + os.sep + file
            if filepath.endswith(".xlsx") and not file.startswith('~'):

                # To open Workbook 
                wb = xlrd.open_workbook(filepath) 
                sheet = wb.sheet_by_index(0) 

                if sheet.cell_value(0, 0) == 'COLLOCATION APPLICATION' and sheet.cell_value(2, 0) == 'SPECIAL PROJECT':
                    
                    site_number =  sheet.cell_value(10, 2)  # ATC Site Number
                    customer.append(int(site_number))
                    customer_name = sheet.cell_value(10,0).upper()
                    customer_equipment = ("CE-" + str(int(site_number)) + "-" + customer_name).strip("(PTY)  LTD")
                    Customer_Equipment_Type = ''
                    
                    customer.append(str(int(site_number)) + "-" + "WT1")# Tower Asset Number 
                    customer.append(customer_equipment) # Customer Equipment Name
                    customer.append("South Africa") # Portfolio Country
                    customer.append("Active") # Status
                    customer.append(customer_name) # Customer Name
                    
                    for i in range(22,28):
                        if sheet.cell_value(i,1) != "":
                            Customer_Equipment_Type = "Ground"
                            ground.append(Customer_Equipment_Type) # Customer Equipment Type
                            ground.append(sheet.cell_value(i,1)) # Ground instalation status
                            ground.append(sheet.cell_value(i,2)) # Ground Equipment Type
                            ground.append(sheet.cell_value(i,4)) # Ground Technology Type       
                            ground.append(sheet.cell_value(i,5)) # Ground Equipment Location
                            ground.append(sheet.cell_value(i,7)) # Ground Equipment Make        
                            ground.append(sheet.cell_value(i,9)) # Ground Equipment Madel
                            ground.append(sheet.cell_value(i,11)) # Ground Lenght        
                            ground.append(sheet.cell_value(i,12)) # Ground width
                            ground.append(sheet.cell_value(i,13)) # Ground hight
                            
                            ground_data = customer + ground
                            cust_eq.writerow(ground_data)
                            print(ground_data)
                        ground.clear()
                    
                    for i in range(32, 52):
                        if sheet.cell_value(i,0) != "":
                            Customer_Equipment_Type = "Tower"
                            antenna.append(Customer_Equipment_Type) # Customer Equipment Type
                            antenna.append(sheet.cell_value(i,1)) #antenna instalation status
                            antenna.append(sheet.cell_value(i,4)) #antenna equipment type
                            antenna.append(sheet.cell_value(i,5)) #antenna mount type
                            antenna.append(sheet.cell_value(i,7)) #antenna make
                            antenna.append(sheet.cell_value(i,8)) #antenna model
                            antenna.append(sheet.cell_value(i,9)) #antenna height
                            antenna.append(sheet.cell_value(i,10)) #antenna width
                            antenna.append(sheet.cell_value(i,11)) #antenna Deep
                            antenna.append(sheet.cell_value(i,12)) #antena azimuth "90/180/270"
   
                            antenna_counter = sheet.cell_value(i,3) #Equipment qty
                            num = 1
                            if (antenna_counter != ''):
                                num = int(antenna_counter)

                            if (sheet.cell_value(i,14) != ''):
                                feeder_qty = int(sheet.cell_value(i,14))
                                antenna.append(int(feeder_qty / num)) #feeder qty
                            else:
                                antenna.append(sheet.cell_value(i,14)) #feeder qty
                            antenna.append(sheet.cell_value(i,15)) #feeder type
                              
                            for y in range(num):
                                tower_data = customer + antenna
                                cust_eq.writerow(tower_data)
                                print(tower_data)
                            antenna.clear()           
                    customer.clear()

                    
                if sheet.cell_value(0, 0) == 'COLLOCATION APPLICATION' and sheet.cell_value(2, 2) != 'Fibre Application' and sheet.cell_value(2, 0)== 'ATC SALES REPRESENTATIVE':
                    
                    site_number =  sheet.cell_value(8, 2)  # ATC Site Number
                    customer.append(int(site_number))
                    customer_name = sheet.cell_value(6,2).upper()
                    customer_equipment = ("CE-" + str(int(site_number)) + "-" + customer_name).strip("(PTY)  LTD")
                    Customer_Equipment_Type = ''
                    
                    customer.append(str(int(site_number)) +"-WT1") #Tower Assets Number
                    customer.append(customer_equipment) # Customer Equipment Name
                    customer.append("South Africa") # Portfolio Country
                    customer.append("Active") # Status
                    customer.append(customer_name) # Customer Name
                    
                    for i in range(3,9):
                        if sheet.cell_value(i,1):
                            Customer_Equipment_Type = "Ground"
                            ground.append(Customer_Equipment_Type) # Customer Equipment Type
                            ground.append(sheet.cell_value(i,1)) # Ground instalation status
                            ground.append(sheet.cell_value(78,i)) # Ground Equipment Type
                            ground.append(sheet.cell_value(82,i)) # Ground Technology Type       
                            ground.append(" ") # Ground Equipment Location
                            ground.append(sheet.cell_value(81,i)) # Ground Equipment Make        
                            ground.append(sheet.cell_value(81,i)) # Ground Equipment Madel
                            ground.append(" ") # Ground Lenght        
                            ground.append(" ") # Ground width
                            ground.append(" ") # Ground hight

                            ground_data = customer + ground
                            cust_eq.writerow(ground_data)
                            print(ground_data)
                            ground.clear()
                    
                    for i in range(3, 9):
                        if sheet.cell_value(87,i) != "":
                            Customer_Equipment_Type = "Tower"
                            antenna.append(Customer_Equipment_Type) # Customer Equipment Type
                            antenna.append(sheet.cell_value(88,i)) # antenna instalation status
                            antenna.append(sheet.cell_value(87,i)) # antenna equipment type
                            antenna.append(sheet.cell_value(91,i)) # antenna mount type
                            antenna.append(sheet.cell_value(92,i)) # antenna make
                            antenna.append(sheet.cell_value(93,i)) # antenna model

                            dimensions = sheet.cell_value(94,i).strip("mm")
                            dimen = dimensions.split("x")              

                            if (len(dimen) == 2):
                                antenna.append(dimen[0])
                                antenna.append(dimen[1])
                            elif(len(dimen) == 3):
                                antenna.append(dimen[0]) # antenna height
                                antenna.append(dimen[1]) # antenna width
                                antenna.append(dimen[2]) # antenna deep
                            else:
                                antenna.append(dimen)
                            antenna.append(sheet.cell_value(97,i)) #antena azimuth "90/180/270"

                            antenna_counter = sheet.cell_value(96,i) # Equipment qty
                            num = 1
                            if (antenna_counter != ''):
                                num = int(antenna_counter)

                            if (sheet.cell_value(103,i) != '' and sheet.cell_value(103,i).isnumeric()):
                                feeder_qty = int(sheet.cell_value(103,i))
                                antenna.append(int(feeder_qty / num)) #feeder/Line qty
                            else:
                                antenna.append(sheet.cell_value(103,i)) #feeder/Line qty

                            antenna.append(sheet.cell_value(105,i)) #feeder/Line type
                            
                            for y in range(num):
                                tower_data = customer + antenna
                                cust_eq.writerow(tower_data)
                                print(tower_data)
                            antenna.clear()
                    customer.clear()                
                                        
print("done!")
                    
```

```python
import os
import xlrd 
import csv

import glob
import time


#Root directory
rootdir = 'C:/OTM'

lates_files = []


def findlatedLoadingSheet(dir):
    
    for subdir, dirs, files in os.walk(rootdir):
        if subdir.endswith("ATC Loading Sheets"):
            for subentry in os.listdir(subdir):
                if subentry.endswith(".xlsx"):
                    subentrypath = os.path.abspath(os.path.join(subdir, subentry))
                    mtime = os.stat(subentrypath).st_mtime
                    max_mtime = 0
                    if mtime > max_mtime:
                        max_mtime = mtime
                        max_dir = subdir
                        max_file = subentry
            lates_file = max_dir + "/" + max_file  
            lates_files.append(lates_file)
#             print (lates_file)
    return (lates_files)
                
findlatedLoadingSheet(rootdir)
```
