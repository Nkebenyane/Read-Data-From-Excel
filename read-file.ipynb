```python
import os
import xlrd 
import csv

eq_header = ["Site Number",
                "Tower Asset Number",
                "Project Number",
                "Project Name",
                "Customer Equipment Number",
                "Customer Equipment Name",
                "Portfolio Country",
                "Status",
                "Customer Equipment Type",
                "Customer Equipment Status",
                "Customer Name",
             
                "Ground Equipment Type",
                "Ground Technology Type",
                "Ground Equipment Location",
                "Ground Equipment Make",
                "Ground Equipment Model",
                "Ground Lenght",
                "Ground Widht",
                "Ground Hight",
             
                "antenna instalation status",
                "antenna equipment type",
                "antenna mount type",  
                "antenna make",  
                "antenna model",  
                "antenna height",             
                "antenna width",  
                "antenna depth",
                "azimuth",
                "feeder/Line qty",
                "feeder/Line type"]


data = ''
ground = []
antenna = []
customer = []

site_number = ''
customer_name

# Auto increment the CE Number:
CE_Number = 1
ground_counter = 0


#Root directory
rootdir = 'C:/OTM'


# run the vbs scipt to encode the excel files
os.system("MyScript.vbs 1")


# Reading form def source and writing in customer_eg file
with open('customer_eq.csv',  'w', newline='') as csvfile:
    cust_eq = csv.writer(csvfile, delimiter=',',escapechar = ' ', quoting=csv.QUOTE_NONE) 
    cust_eq.writerow(eq_header)
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            filepath = subdir + os.sep + file
            if filepath.endswith(".xlsx") and not file.startswith('~'):

                # To open Workbook 
                wb = xlrd.open_workbook(filepath) 
                sheet = wb.sheet_by_index(0) 

                if sheet.cell_value(0, 0) == 'COLLOCATION APPLICATION' and sheet.cell_value(2, 0) == 'SPECIAL PROJECT':
                    
                    antenna_counter = sheet.cell_value(32,3)
                    num = 1
                    if (antenna_counter != ''):
                        num = int(antenna_counter)
                    
                    site_number =  sheet.cell_value(10, 2)  # ATC Site Number
                    custumer_name = sheet.cell_value(10,0)
                    customer.append(int(site_number))
                    
                    customer.append(str(int(site_number)) + "-" + "WT1")# Tower Asset Number 
                    customer.append(" ") #Project Number
                    customer.append(" ") # Project Name
                    customer.append(" ") # Customer Equipment Number
                    customer.append("CE" + "-" + str(int(site_number)) + "-" + customer_name) # Customer Equipment Name
                    customer.append("South Africa") # Portfolio Country
                    customer.append("Active") # Status
                    customer.append("Tower") # Customer Equipment Type
                    customer.append("Contract") # Customer Equipment Stutas
                    customer.append(customer_name) # Customer Name
                    
                    for i in range(22,28):
                        if sheet.cell_value(i,1) != "":
                            ground.append(sheet.cell_value(i,2)) # Ground Equipment Type
                            ground.append(sheet.cell_value(i,4)) # Ground Technology Type       
                            ground.append(sheet.cell_value(i,5)) # Ground Equipment Location
                            ground.append(sheet.cell_value(i,7)) # Ground Equipment Make        
                            ground.append(sheet.cell_value(i,9)) # Ground Equipment Madel
                            ground.append(sheet.cell_value(i,11)) # Ground Lenght        
                            ground.append(sheet.cell_value(i,12)) # Ground width
                            ground.append(sheet.cell_value(i,13)) # Ground hight
                            
                            data = customer + ground
                        ground.clear()
                    customer.clear()                
                    
                    for i in range(32, 52):
                        if sheet.cell_value(i,0) != "":   
                            antenna.append(sheet.cell_value(i,1)) #antenna instalation status
                            antenna.append(sheet.cell_value(i,4)) #antenna equipment type
                            antenna.append(sheet.cell_value(i,5)) #antenna mount type
                            antenna.append(sheet.cell_value(i,7)) #antenna mount make
                            antenna.append(sheet.cell_value(i,8)) #antenna mount model
                            antenna.append(sheet.cell_value(i,9)) #antenna mount height
                            antenna.append(sheet.cell_value(i,10)) #antenna mount width
                            antenna.append(sheet.cell_value(i,11)) #antenna mount Deep
                            antenna.append(sheet.cell_value(i,12)) #antena azimuth "90/180/270"
                            if (sheet.cell_value(i,14) != ''):
                                feeder_qty = int(sheet.cell_value(i,14))
                                antenna.append(int(feeder_qty / num)) #feeder qty
                            else:
                                antenna.append(sheet.cell_value(i,14)) #feeder qty
                            antenna.append(sheet.cell_value(i,15)) #feeder type
                              
                            for y in range(num):
                                print (y)
                                st = data + antenna
                                cust_eq.writerow(st)
                                print(st)  
                            antenna.clear()           
                    customer.clear()

                    
                if sheet.cell_value(0, 0) == 'COLLOCATION APPLICATION' and sheet.cell_value(2, 2) != 'Fibre Application' and sheet.cell_value(2, 0)== 'ATC SALES REPRESENTATIVE':
                    site_number =  sheet.cell_value(8, 2)  # ATC Site Number
                    customer.append(int(site_number))
                    customer_name = sheet.cell_value(6,2)
                    
                    customer.append(str(int(site_number)) +"-WT1") #Towe Assets Number
                    customer.append(" ") #Project Number
                    customer.append(" ") # Project Name
                    customer.append(" ")# Customer Equipment Number
                    customer.append("CE-" + str(int(site_number)) + "-" + customer_name) # Customer Equipment Name
                    customer.append("South Africa") # Portfolio Country
                    customer.append("Active") # Status
                    customer.append("Tower") # Customer Equipment Type
                    customer.append("Contract") # Customer Equipment Stutas
                    customer.append(customer_name) # Customer Name
                    
                    
                    
                    st = customer
                    cust_eq.writerow(st)
                    print(customer)
                    customer.clear()
                        
                
                    
print("done!")
                    
```
